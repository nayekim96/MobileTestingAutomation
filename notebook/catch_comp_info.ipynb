## 캐치 - 기업정보 크롤링

```python
import requests
from bs4 import BeautifulSoup as BS
import json
import pandas as pd
```

```python
url = 'https://www.catch.co.kr/api/v1.0/comp/compMajor/getMainCompanyListV2'

payload = {
    'AreaSido': "",
    'CName': "",
    'CategoryName': "",
    'GroupName': "",
    'IsInHiring': "0",
    'IsNewList': "0",
    'JCode': "",
    'NowPage': "1",
    'PageSize': "7000",
    'Salary': "0,0",
    'Size': "1,3",
    'Sort': "profit",
    'ThemeName': ""
}

head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.3'}
```

```python
r = requests.post(url, json=payload, headers=head)
```

```python
bs = BS(r.text)
```

```python
company_list = json.loads(bs.find('p').text)['companyList']
```

```python
len(company_list)
```

### compId : 기업코드

```python
compId = []
for info in company_list:
    compId.append(info['CompID'])
```
