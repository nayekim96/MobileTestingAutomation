## 캐치 - 대기업/중소기업 전체 면접후기 데이터

### 기업코드 크롤링

```python
# 패키지 import

import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import re
import json
```

```python
# 크롤링 정보
url = 'https://www.catch.co.kr/api/v1.0/comp/compMajor/getMainCompanyListV2'
payload = {
    'AreaSido': "",
    'CName': "",
    'CategoryName': "",
    'GroupName': "",
    'IsInHiring': "0",
    'IsNewList': "0",
    'JCode': "",
    'NowPage': "1",
    'PageSize': "7000",
    'Salary': "0,0",
    'Size': "1,3",
    'Sort': "profit",
    'ThemeName': ""
}
head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.3'}

# post
r = requests.post(url, json=payload, headers=head)
bs = BS(r.text)

# 기업 정보 추출
company_list = json.loads(bs.find('p').text)['companyList']

# 기업코드 추출 - compID
CompanyCode = []
for info in company_list:
    CompanyCode.append(info['CompID'])
```

### 대기업/중소기업 전체 면접 후기 크롤링

```python
from tqdm import tqdm


batch_size = 100
all_reviews = []

for i in tqdm(range(0, len(CompanyCode), batch_size)):
    batch = CompanyCode[i:i+batch_size]

    for compid in batch:
        try:
            url = f'https://www.catch.co.kr/api/v1.0/jobn/interview/comp/reviewList/{compid}?CurPage=1&PageSize=1&JobCode=&RecruitGubun=&IsPass=&ProcessType='
            head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36'}
            r = requests.get(url, headers=head)

            # 면접 후기 총 개수
            PageSize = json.loads(r.text)['totalCount']

            # 면접 후기 전체 데이터 크롤링
            url = f'https://www.catch.co.kr/api/v1.0/jobn/interview/comp/reviewList/{compid}?CurPage=1&PageSize={PageSize}&JobCode=&RecruitGubun=&IsPass=&ProcessType='
            r = requests.get(url, headers=head)
            bs = BS(r.text)

            df = pd.DataFrame(json.loads(bs.find('p').text)['review'])
            all_reviews.append(df)
            
        except Exception as e:
            print(f"Error processing Company ID {compid}: {e}")
```
